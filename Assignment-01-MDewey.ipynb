## Q1: Simulate aliasing of sinusoidal signals

```python
import numpy as np
import matplotlib.pyplot as plt
import math 
%matplotlib inline
```

For this problem you will simulate the phenomena of aliasing.  This will give you pracitice in making synthetic time series, subsetting data, and plotting your results.

**1** To start, make a time vector that goes from 0 to 100 s, with 10000 entries in it.  Then make a synthetic time series that has two close in frequency sine waves:

$$ x = A_1 \cos (\omega t) + A_2 \cos ((\omega+\delta \omega ) t) $$

where $\omega$ is the "carrier" frequency, and $\delta \omega << \omega$ is the difference in frequencies between the two waves.  For simplicty, you could make $A_1=A_2=1.0\ [V]$.  Feel free to choose an interesting $\omega$.

<hr>

```python
t = np.linspace(0.,100.,10000)
A1 = 1.0   #[V]
A2 = 1.0   #[V]
dw = 0.1
w = 1.
f1 = A1*np.cos(w*t)
f2 = A2*np.cos((w+dw)*t)
x = f1+f2
```

**2** Now plot the two component waves and the full value of $x$ as a time series.  Preferably on the same axis.  Make sure to label your axes and include a legend<hr>

```python
%matplotlib inline
fig = plt.figure(figsize=(11,4))
ax = fig.add_subplot(1,1,1)
ax.plot(t,f1, label='Signal 1')
ax.plot(t,f2, label='Signal 2 (offset by dw = 0.1)')
ax.plot(t,x, label='Sum of signals' )
ax.set_xlabel('Time [s]'),ax.set_ylabel('Data'),ax.legend()
ax.set_title('Sum of two component waves')
```

**3** Try a few different values of $\delta\omega$, and compare.<hr>

```python
dw = 0.01
f1 = A1*np.cos(w*t)
f2 = A2*np.cos((w+dw)*t)
x = f1+f2
fig = plt.figure(figsize=(11,5))
ax = fig.add_subplot(3,1,1)
ax.plot(t,f1,label='dw=0.01')
ax.plot(t,f2)
ax.plot(t,x)
ax.legend()

dw = 0.1
f1 = A1*np.cos(w*t)
f2 = A2*np.cos((w+dw)*t)
x = f1+f2
ax2 = fig.add_subplot(3,1,2)
ax2.plot(t,f1, label='dw=0.1')
ax2.plot(t,f2)
ax2.plot(t,x)
ax2.legend()

dw = 0.5
f1 = A1*np.cos(w*t)
f2 = A2*np.cos((w+dw)*t)
x = f1+f2
ax3 = fig.add_subplot(3,1,3)
ax3.plot(t,f1, label='dw=0.5')
ax3.plot(t,f2)
ax3.plot(t,x)
ax3.set_xlabel('Time [s]')
ax3.legend()
ax.set_title('Time series with varying frequency differences')
```

## Q2: Demonstrate the central limit theorem for a very unfair coin flip.

Consider a coin that is twice as likely to come up heads as tails. To put a number on it, call tails $x=-1/2$ and heads $x=+1/2$.  

**1** What are the mean, $\mu_x$, and variance $\sigma_x^2$ of this process?

<hr>

 For a discrete variable:
 
 $\mu_x$ = $\sum$$x$$f(x)$   where f(x) is the pdf
 
 $\sigma_x^2$ = $\sum$$(x-\mu_x)^2$$f(x)$
 
 Therefore, for a single coin flip:
 
 $\mu_x$ = $(0.5)(0.5)+(-0.5)(0.5)$ = 0
 
 $\sigma_x^2$ = $(0.5)^2(0.5)+(-0.5)^2(0.5)$ = 0.25

**2** Plot or represent the *probability distribution*  (or cumulative distrubution) of this process.
<hr>

```python
x = np.arange(-1.,1.,0.01)
a=-0.5
b=0.5
p = 0.*x
p[(x>a)]=0.5
p[(x>b)]=1.
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
ax.plot(x,p)
ax.set_xlabel('x - heads or tails');ax.set_ylabel('p(x)');ax.set_ylim([-0.1,1.1])
ax.set_title('Probabilty Distribution for unbiased coin flip')
```

**3** Write a function using the `np.random.norm` function to simulate the coin flip.  It shoudl return $+1/2$ twice as often as $-1/2$.  Demonstrate that your function works.
<hr>

```python
def getflip(N,doprint=False,theseed=None):
    import numpy as np
    import scipy.stats as stats
    np.random.seed(seed=theseed)
    xx=np.random.rand(N)
    H=0.
    T=0.
    for i in range(N):
        if xx[i]<0.6666666:
            xx[i]=0.5
            H=H+1.
        else:
            xx[i]=-0.5
            T=T+1.
    if doprint:
        print 'Number of flips: %d' % N
        print 'Heads: %d' %H
        print 'Tails: %d' %T
        print 'Mean: %1.3f'%np.mean(xx)
        print 'Variance: %1.3f'%np.var(xx)
        print 'Skewness: %1.3f'%stats.skew(xx)
        print 'Kurtosis: %1.3f'%stats.kurtosis(xx)
        print ''
    return xx
xx=getflip(200,doprint=True,theseed=789)
```

**4** Perform $N_{flip}$ and take the sample mean $\overline{x}$ of the values.  Do this 2000 times, and show that as $N_{flip}\to\infty$ the distribution of $\overline{x}$ approaches a Normal distribution by plotting the histograms of the means for each.  For $N_{roll}$ choose 2, 16, 64, 256. 

Comment on how the convergence of $\overline{x}$ for the unfair coin compares to the convergence of the sample mean of the dice roll.  
<hr>

```python
Ntrials = 2000
means=np.zeros(Ntrials)
for i in range(Ntrials):
    xx=getflip(2)
    means[i]=np.mean(xx)
means2=means
fig=plt.figure(figsize=(11,2))
ax=fig.add_subplot(1,1,1)
ax.hist(means,30)
ax.set_xlabel('$\overline{x}\ \ N=2$')
ax.set_ylabel('Occurences')

means=np.zeros(Ntrials)
for i in range(Ntrials):
    xx=getflip(16)
    means[i]=np.mean(xx)
means16=means
fig2=plt.figure(figsize=(11,2))
ax2=fig2.add_subplot(1,1,1)
ax2.hist(means,30)
ax2.set_xlabel('$\overline{x}\ \ N=16$')
ax2.set_ylabel('Occurences')

means=np.zeros(Ntrials)
for i in range(Ntrials):
    xx=getflip(64)
    means[i]=np.mean(xx)
means64=means
fig3=plt.figure(figsize=(11,2))
ax3=fig3.add_subplot(1,1,1)
ax3.hist(means,30)
ax3.set_xlabel('$\overline{x}\ \ N=64$')
ax3.set_ylabel('Occurences')

means=np.zeros(Ntrials)
for i in range(Ntrials):
    xx=getflip(256)
    means[i]=np.mean(xx)
means256=means
fig4=plt.figure(figsize=(11,2))
ax4=fig4.add_subplot(1,1,1)
ax4.hist(means,30)
ax4.set_xlabel('$\overline{x}\ \ N=256$')
ax4.set_ylabel('Occurences')
```

The fair dice roll converged to a normal distribution about zero, but with the unfair coin the distribution converged to a normal distribution about 0.1667 (two-thirds of the distance between -0.5 and 0.5, indicating 0.5 is twice as favoured as -0.5)

```python

m2=np.mean(means2)
st2=np.std(means2)
normd2=((means2-m2)/st2)

m16=np.mean(means16)
st16=np.std(means16)
normd16=((means16-m16)/st16)

m64=np.mean(means64)
st64=np.std(means64)
normd64=((means64-m64)/st64)

m256=np.mean(means256)
st256=np.std(means256)
normd256=((means256-m256)/st256)

fig=plt.figure(figsize=(11,4))
ax=fig.add_subplot(1,1,1)
bins=30
ax.hist(normd2,bins,histtype='step',normed=True,label='2 flips')
ax.hist(normd16,bins,histtype='step',normed=True,label='16 flips')
ax.hist(normd64,bins,histtype='step',normed=True,label='64 flips')
ax.hist(normd256,bins,histtype='step',normed=True,label='256 flips')

from scipy.stats import norm
range = np.arange(-3, 3, 0.001)
ax.plot(range, norm.cdf(range,0,1.),'k--',label='CDF Normal Distribution')
ax.legend()
ax.set_title('Means removed, normalized by standard deviation')
ax.set_xlabel('$\overline{x}\ \ N$')
ax.set_ylabel('Occurances')
```
