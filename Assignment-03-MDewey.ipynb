````python
import numpy as np
import matplotlib.pyplot as plt
import math 
import scipy.stats as stats
import cmath
%matplotlib inline
````

````python
def lcc(dat,t):
    good = np.isfinite(dat)
    dat2 = dat[good]
    datmean = np.mean(dat2)
    dat4 = dat-datmean
    notgood = np.isnan(dat)
    dat4[notgood]=0

    fig,bx=plt.subplots(1,1)
    bx.plot(dat4)
    bx.set_ylabel('Temperature [$^\circ C$]')
    bx.set_xlabel('Time [hr]')
    bx.set_title('Normalized temperature')

    lags = np.arange(-(24.*t),(24.*t))
    cxx = 0.*lags
    for ind, tau in enumerate(lags):
        if tau<0:
            cxx[ind]=np.mean(dat4[:tau]*dat4[-tau:])
        elif tau==0:
            cxx[ind]=np.mean(dat4*dat4)
        else:
            cxx[ind]=np.mean(dat4[:-tau]*dat4[tau:])
    cxx=cxx/np.mean(dc4*dc4)

    fig,ax=plt.subplots(1,1)
    ax.plot(lags,cxx)
    ax.set_xlabel(r'$\tau [hr]$')
    ax.set_ylabel(r'$\rho_{xx} [V^2]$')
    ax.set_title('Lag Correlation')

lcc(dc,30)
````

You can see a daily cycle in the correlation, measurements made at the same time of day are well correlated, whereas measurements made 12 hours apart are not. You can also see overall less correlation as the lag increases. This means data collected at the beginning of the month is not very well correlated with data collected at the end of the month. If you entended the lag out farther than 30 days, you may see a yearly cycle, (ie May 2011 being well correlated to May 2012 but not September 2011).

## Q2 Emperically show that the expected value of a finite spectrum approaches the true spectrum as T goes to infinity.

**1** Consider the process $x(t) = \int_{t-b}^t y(t) \ \mathrm{dt}$ where $y(t)$ is uncorrelated white noise so that $R_{yy}(\tau)=\delta(\tau)$.  Compute $R_{xx}(\tau)$.

$R_{xx} = E[x(t)x(t+\tau)]$
       $= E[\int_{t-b}^t y(t')dt' \int_{t+\tau-b}^{t+\tau}y(t'')dt'']\:\:$
       $= \int_{t-b}^t \int_{t+\tau-b}^{t+\tau} E[y(t')y(t'')]dt''dt'\:\:$
       $= \int_{t-b}^t \int_{t+\tau-b}^{t+\tau} \delta(t''-t')dt''dt'\:\:$
       $= \int_{t-b}^t [ 1\:if\: t+\tau-b<t'\:<\:t+\tau,\:0\:else] dt'\:\:$
       $= \int_{t-b}^{t+\tau-b} \: 0 \: dt' + \int_{t+\tau-b}^t dt'\:\:$
       $= t-(t+\tau-b) = b-\tau$
       
In order to normalize $R_{xx}$, (have it equal 1 when $\tau = 0$), then you divide by b:

$R_{xx} = (b-\tau)/b$

**2** Show emperically that you have the right *lag auto correlation* by computing the lag correlation of a random discrete timeseries $x(t)$.  Hint: have a look at `np.convolve` to do the integration. Hint 2: Make sure your time series is long enough to get a nice representation of $R_{xx}(\tau)$.  Show what happens if N is too small and what happens if it gets larger.
(Hint to do this well, I needed N to get pretty large, like 5e6,  I used `N in [500, 5000, 50000, 5e6]`, but my machine has a good amount of memory.  Don't *test* on the bigger values, just add them for the final product).

````python
def Rxx(N,b,t,navg):
    Y = np.random.randn(N)
    dn=N/navg
    CX = np.zeros(t)
    for j in range(navg):
        x = np.zeros(dn-b)
        y = Y[j*dn:(j+1)*dn]
        for i in range(dn-b):
            x[i] = np.sum(y[i:i+b])

        lags = np.arange(0,t)
        cx = 0.*lags
        for ind, tau in enumerate(lags):
            if tau<0:
                cx[ind]=np.mean(x[:tau]*x[-tau:])
            elif tau==0:
                cx[ind]=np.mean(x*x)
            else:
                cx[ind]=np.mean(x[:-tau]*x[tau:])
        cx=cx/np.mean(x*x)
        CX = CX + cx
    cx = CX/navg
    sN = str(N)
    return sN,lags,cx

N1,lags1,cx1 = Rxx(500,10,50,1)
N2,lags2,cx2 = Rxx(5000,10,50,1)
N3,lags3,cx3 = Rxx(50000,10,50,1)
N4,lags4,cx4 = Rxx(5000000,10,50,1)

    
fig,ax=plt.subplots(1,1)
ax.plot(lags1,cx1,label=N1)
ax.plot(lags2,cx2,label=N2)
ax.plot(lags3,cx3,label=N3)
ax.plot(lags4,cx4,label=N4)
ax.legend()
ax.set_xlabel(r'$\tau [s]$')
ax.set_ylabel(r'$R_{xx} [V^2]$')
ax.set_title('Lag Auto Correlation (b=10)')
fig.show()
    
````

````python
#dt=1s, f = [0:1/2]Hz (frequency goes from 0 to 0.5 Hz in step sizes 1/n where n is the length of the time series)

def dft(x,dt):
    n = np.size(x)
    nf = n/2+1
    i = np.arange(0,n)
    F = np.arange(1,nf+1,dtype=complex)
    FT = np.arange(1,nf+1,dtype=float)
    f = np.arange(1,nf+1,dtype=float)
    for k in range(0,nf):
        f[k] = k/(n*dt)
        a = 2*np.pi*f[k]
        F[k] = sum(x*np.exp(-cmath.sqrt(-1)*a*i))
        FT[k] = np.sqrt(((np.real(F[k]))*(np.real(F[k])))+((np.imag(F[k]))*(np.imag(F[k]))))
    return FT,f

FT1,f1 = dft(cx1,1.)
FT2,f2 = dft(cx2,1.)
FT3,f3 = dft(cx3,1.)
FT4,f4 = dft(cx4,1.)
fig,ax = plt.subplots(1,1)
ax.plot(f1,FT1,label='500')
ax.plot(f2,FT2,label='5000')
ax.plot(f3,FT3,label='50000')
ax.plot(f4,FT4,label='5e6')
ax.legend()
fig.show()
ax.set_yscale('log')
ax.set_xlabel('f [Hz]')
ax.set_ylabel('Log($S_{xx}$) [$V^2$/Hz]')
````
