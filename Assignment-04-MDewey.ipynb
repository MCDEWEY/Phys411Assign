````python
def dft(x,dt):   ##note: dt must be a float, or f will be calculated with int. division and will be all zeros
    n = np.size(x)
    nf = n/2+1
    i = np.arange(0,n)
    F = np.arange(1,nf+1,dtype=complex)
    FT = np.arange(1,nf+1,dtype=float)
    f = np.arange(1,nf+1,dtype=float)
    for k in range(0,nf):
        f[k] = k/(n*dt)
        a = 2*np.pi*f[k]
        F[k] = sum(x*np.exp(-cmath.sqrt(-1)*a*i))
        FT[k] = np.sqrt(((np.real(F[k]))*(np.real(F[k])))+((np.imag(F[k]))*(np.imag(F[k]))))
    return FT,dt,f,F
````

**1** Write a short function to create a "red" time series by taking the cumulative sum of Normally distributed random data, and plot one realization.  The number of data points $N$ can be relatively small (say 1000).

````python
def red(N):
    x = np.cumsum(np.random.randn(N))
    fig,ax = plt.subplots(1,1)
    ax.plot(x)
    ax.set_title('"red" Time Series')
    ax.set_xlabel('t [s]')
    ax.set_ylabel('x [V]')
    return x
    
x = red(1000)
````

**2** Write a second function to return the estimate of the power spectrum of $x$.  For posterity sake, include a value of the sampling frequency, and return the proper frequencies for the estimate.  Plot the resulting spectrum.

````python
def pspec(x,dt):
    FT,dt,f,F = dft(x,dt)
    n = np.size(F)
    T = n*dt
    Gxx = (2/T)*(F*np.conjugate(F))
    return Gxx,dt,f,F

Gxx,dt,f,F = pspec(x,1.)
fig,ax = plt.subplots(1,1)
ax.plot(f,Gxx)
ax.set_yscale('log')
ax.set_xscale('log')
ax.set_xlabel('f [Hz]')
ax.set_ylabel('Gxx [$V^{2}/Hz$]')
ax.set_title('Power Spectrum of red noise')
````
