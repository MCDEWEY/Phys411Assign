### Error bounds on coherence estimates

<div style='background:#F0F0F0'>  The 95% confidence intervals for the coherence squared $\gamma^2_{xy}(f)$ between two variables $x(t)$ and $y(t)$ are approximately:
$$ \gamma^2_{xy} \pm \frac{\sqrt{2}\left(1+\gamma^2_{xy}\right)}{\left|\gamma_{xy} \right|\sqrt{n_d}}$$
where $n_d$ is the number of blocks used to calculate the coherence.  

You **may** use `matplotlib.mlab.csd` for the following question, but you had best use it correctly!
</div>

<div style='background:#F0F0F0'>**1** Make an input process $x(t)$ as random Gaussian noise, and the output process the sum of a noise source and a running average of the input: $y(t)=n(t)+\frac{1}{b}\int_{t-b}^t x(t')dt'$ and plot (Zoom so we can see whats going on, and label your plot).  Make sure you generate a large time series.</div> 

```python
import numpy as np
import matplotlib.mlab as mlab
import scipy.stats as stats
import matplotlib.pyplot as plt
import math 
import cmath
%matplotlib inline
```

```python
N=10000
window = 5.0
x = np.random.randn(N)
weights = np.repeat(1.0,window)/window
ave=np.convolve(x,weights,'same')
y=np.random.randn(N)/20.+ave
```

```python
fig,ax = plt.subplots(1,1)
fig.set_size_inches(15,4)
ax.plot(x,label='Gaussian input white noise, x(t)')
ax.plot(y,'g',label='Output, y(t)')
ax.set_title('Full Time Series')
ax.set_xlabel('t [s]')
ax.legend()
```

```python
fig,ax = plt.subplots(1,1)
fig.set_size_inches(15,4)
ax.plot(x[2000:2500],label='Gaussian input white noise, x(t)')
ax.plot(y[2000:2500],'g',label='Output, y(t)')
ax.set_title('Time Series from t=2000s to t=2500s')
ax.set_xlabel('t [s]')
ax.legend()
a=ax.get_xticks().tolist()
for idx,val in enumerate(a):
    a[idx]+=2000
ax.set_xticklabels(a)
```

<div style='background:#F0F0F0'>**2** Compute the coherence squared of the input and the output.  Make a plot with the spectra in the upper panel and the coherence squared in the middle panel and the phase in the bottom.

You will need to choose a value for your fft length.  Choose a relative low one to get a smooth plot.
</div>

```python
def coherenceplt(nfft):
    dt=1.
    cFx,f=mlab.csd(x,y,NFFT=nfft,Fs=1./dt,window=mlab.window_hanning,noverlap=nfft/2)
    px,fxn=mlab.psd(x,NFFT=nfft,Fs=1./dt,window=mlab.window_hanning,noverlap=nfft/2)
    py,fxn=mlab.psd(y,NFFT=nfft,Fs=1./dt,window=mlab.window_hanning,noverlap=nfft/2)
    Gamma2 = abs(cFx)**2/py/px

    fig = plt.figure(figsize=(12,12))
    bx = fig.add_subplot(3,1,1)
    bx.plot(y)
    bx.set_ylabel('y(t)')
    bx.set_xlabel('t [s]')
    bx.set_title(' Signal, coherence squared, and phase for NFFT=%1.1f' %nfft)
    ax = fig.add_subplot(3,1,2)
    ax.semilogx(f,Gamma2)
    inter = (np.sqrt(2)*(1+Gamma2))/(np.abs(Gamma2)*((2*N/nfft)-1))
    ax.fill_between(f,Gamma2+inter,Gamma2-inter,alpha=0.5,
                linewidth=0.0,edgecolor=None,color=None,facecolor='b')
    ax.set_ylim((0,1.1))
    ax.set_ylabel(r'$\gamma^2$')
    ax.set_xlabel('f [$s^{-1}$]')
    cx = fig.add_subplot(3,1,3)
    ang = np.angle(cFx)
    cx.semilogx(f,ang/np.pi)
    cx.set_ylabel('$Phase / \pi$')
    cx.set_xlabel('f [s]')
    
nfft = np.size(x)/20
coherenceplt(nfft)
```

```python
nfft1 = np.size(x)/50
coherenceplt(nfft1)
nfft2 = np.size(x)/10
coherenceplt(nfft2)
nfft3 = np.size(x)/2
coherenceplt(nfft3)
```

For the first two NFFT sizes, both peaks in $\gamma^2$ are significant. 
<div style='background:#F0F0F0'>Download the data abaove. `time` is a measurement time in seconds. `shear` is a little less than 300 s of data from a turbulence shear probe. Units are $s^{-1}$. `ac` is a suitably scaled acceleration of the vehicle containing the shear probes.

**1** Plot the shear time series and the accelerometer time series.  Comment on any similarities and differences.  Comment on the stationarity of the time series.

```python
fig = plt.figure(figsize=(12,12))
ax= fig.add_subplot(2,1,1)
ax.plot(time,shear)
ax.set_ylabel('Turbulent Shear [$s^{-1}$]')
ax.set_xlabel('Time [s]')
ax.set_title('Turbulent Shear')
bx = fig.add_subplot(2,1,2)
bx.plot(time,ac)
bx.set_ylabel('Scaled acceleration [$m/s^2$]')
bx.set_xlabel('Time [s]')
bx.set_title('Scaled acceleration of shear probe')
```

There seems to be greater turbulent shear when the probe is vibrating strongly (when the acceleration is varying greatly). 
The scaled acceleration seems to be stationary, as it does not depend much on time (althought there are greater variations in acceleration towards the end of the time series). The shear is less stationary, as there is a significant change in its behaviour later in the timeseries (the effect of the greater variation in acceleration is enhanced in the shear).

<div style='background:#F0F0F0'> **2** Compute the power spectra of the two signals and the coherence squared (you could do the phase too, but it is not too thrilling).  Comment on the spectral similarities and the coherence.  

```python
dt=1.
nfft = np.size(time)/50
cF,f=mlab.csd(ac,shear,NFFT=nfft,Fs=1./dt,window=mlab.window_hanning,noverlap=nfft/2)
pac,fxn=mlab.psd(ac,NFFT=nfft,Fs=1./dt,window=mlab.window_hanning,noverlap=nfft/2)
psh,fxn=mlab.psd(shear,NFFT=nfft,Fs=1./dt,window=mlab.window_hanning,noverlap=nfft/2)
Gamma2 = abs(cF)**2/pac/psh
```

```python
fig,ax = plt.subplots(1,1)
ax.semilogx(f,Gamma2)
ax.set_ylim((0,1.1))
ax.set_ylabel(r'$\gamma^2$')
ax.set_xlabel('f [$s^{-1}$]')
fig.set_size_inches(12,4)
ax.set_title('Coherence Squared')

fig,bx = plt.subplots(1,1)
bx.loglog(fxn,pac,label='Probe vibration')
bx.loglog(fxn,psh,label='Shear')
bx.set_xlabel('f [$s^{-1}$]')
bx.set_ylabel('Power Spectra Density $[s^{-2} Hz^{-1}]$')
bx.legend()
bx.set_title('Power Spectra')
fig.set_size_inches(12,4)
```

Two peaks in the shear can clearly be seen to be dependent on the probe vibration noise, at $10^{-1}$Hz and 0.3Hz. This can also be seen the the coherence, where the influence on shear by the vibration is evident in the two large peaks at the same frequencies. At these frequencies the shear is almost entirely (80% and >90% respectively) due to the vibrations, where as at all other frequencies the acceleration has little effect on the shear.

<div style='background:#F0F0F0'> **3** Compute a spectra that has the contamination from the vibration removed using the linear technique described in the notes. Comment on how well your correction worked.

```python
G = psh-abs(cF)**2/pac
fig,ax = plt.subplots(1,1)
ax.loglog(f,pac,label='Probe vibration')
ax.loglog(f,psh,label='Shear')
ax.loglog(f,G,linewidth=1.5,label='Shear with vibration removed')
ax.set_xlabel('f [$s^{-1}$]')
ax.set_ylabel('Power Spectra Density $[s^{-2} Hz^{-1}]$')
ax.legend()
fig.set_size_inches(15,6)
```

The correction worked very well for the peak at 0.1Hz, it is entirely removed and the general trend in the shear seems ot be preserved. The peak at 0.3Hz is mostly removed, althought there is still a significant peak.

```python
tlist = time.tolist()

qmin = tlist.index(50.0032)
qmax = tlist.index(100.0)
tmin = tlist.index(200.0)
tmax = tlist.index(250.0032)

acq = ac[qmin:qmax]
shq = shear[qmin:qmax]
act = ac[tmin:tmax]
sht = shear[tmin:tmax]

dt=1.
nfft = np.size(acq)/50
cFq,fq=mlab.csd(acq,shq,NFFT=nfft,Fs=1./dt,window=mlab.window_hanning,noverlap=nfft/2)
pacq,f=mlab.psd(acq,NFFT=nfft,Fs=1./dt,window=mlab.window_hanning,noverlap=nfft/2)
pshq,f=mlab.psd(shq,NFFT=nfft,Fs=1./dt,window=mlab.window_hanning,noverlap=nfft/2)
Gq = pshq-abs(cFq)**2/pacq

fig,ax = plt.subplots(1,1)
ax.loglog(fq,pacq,label='Probe vibration')
ax.loglog(fq,pshq,label='Shear')
ax.loglog(fq,Gq,linewidth=2.0,label='Shear with vibration removed')
ax.set_xlabel('f [$s^{-1}$]')
ax.set_ylabel('Power Spectra Density $[s^{-2} Hz^{-1}]$')
ax.legend()
ax.set_title('"Quiet Period" Correction')
fig.set_size_inches(15,6)

cFt,ft=mlab.csd(act,sht,NFFT=nfft,Fs=1./dt,window=mlab.window_hanning,noverlap=nfft/2)
pact,f=mlab.psd(act,NFFT=nfft,Fs=1./dt,window=mlab.window_hanning,noverlap=nfft/2)
psht,f=mlab.psd(sht,NFFT=nfft,Fs=1./dt,window=mlab.window_hanning,noverlap=nfft/2)
Gt = psht-abs(cFt)**2/pact

fig,ax = plt.subplots(1,1)
ax.loglog(ft,pact,label='Probe vibration')
ax.loglog(ft,psht,label='Shear')
ax.loglog(ft,Gt,linewidth=2.0,label='Shear with vibration removed')
ax.set_xlabel('f [$s^{-1}$]')
ax.set_ylabel('Power Spectra Density $[s^{-2} Hz^{-1}]$')
ax.legend()
ax.set_title('"Turbulent Period" Correction')
fig.set_size_inches(15,6)
```

The correction seems to do a better job in the period which seemed turbulent. This may be because in this area, most of the variation in shear is due to the vibrations, so the corrected spectra is very similar to the original output, minus obvious peaks due to vibration noise.
In the period which seemed quiet, the correction did not do such a good job. This may be because variations in the period are not due to vibrations alone.
