### **Q1** Frequency-domain filtering

<div style='background:#F0F0F0'>  Consider a time series with a sharp jump in the middle as in the notes:
</div>

```python
import numpy as np
import matplotlib.mlab as mlab
import scipy.stats as stats
import matplotlib.pyplot as plt
import math 
import cmath
import scipy.signal as signal
%matplotlib inline
```

```python
N=1001
dt=1.
t=np.arange(0,N*dt,dt)

x = np.cumsum(np.random.randn(N)) # red noise
f0=0.03 # Hz
x = x+ 40.*np.sin(f0*2*np.pi*t) # + sine wave
x=x-np.mean(x)
noise = 60.*np.random.randn(N)
xn = x+noise
xn[N/2:]=xn[N/2:]-2000.
# remove the mean!!
xn = xn-np.mean(xn)
```

<div style='background:#F0F0F0'>**1** Implement a low-pass filter of the data in the frequency domain, with a cutoff frequency of $f_c=0.06\ Hz$.</div>
   
  1. plot the signal and the smoothed signal (suggest zooming in on the step).
  2. plot the spectral response $|H(f)|$
  3. plot the phase response $\mathrm{angle}(H(f))$.

<div style='background:#F0F0F0'>
Be sure to comment on how good your filter is.
</div>

```python
fc = 0.06
nfft = N/10
X = np.fft.fft(xn)
N = np.shape(xn)[0]
ff = np.linspace(0,1./dt,N)
X[(ff>fc) & (ff<(1./dt-fc))]=1.e-15
y = np.real(np.fft.ifft(X))+1.e-17*np.random.randn(N)
py, fy = mlab.psd(y,NFFT=nfft,Fs=1./dt)
px, fx = mlab.psd(xn,NFFT=nfft,Fs=1./dt)
pn, fn = mlab.psd(noise,NFFT=nfft,Fs=1./dt)
Gxy, f = mlab.csd(xn,y,NFFT=nfft,Fs=1./dt)

fig,ax = plt.subplots(3,1)
fig.set_size_inches(12,12)
ax[0].plot(t[400:600],xn[400:600], label='signal')
ax[0].plot(t[400:600],y[400:600], label='smoothed')
ax[0].set_xlabel('t [s]')
ax[0].set_ylabel('x [V]')
ax[0].legend()
ax[0].set_title('Boxcar')

ax[1].loglog(fn,np.sqrt(py/(px+pn)),'g')
ax[1].set_ylabel('|H(f)|')
ax[1].set_xlabel('f [Hz]')

ax[2].semilogx(f,np.angle(Gxy),'g')
ax[2].set_ylabel('phase')
ax[2].set_xlabel('f [Hz]')
```

The higher frequencies are filtered, but it drops off quite slowly after 0.06Hz, so we still see effects of higher frequencies in the filtered signal.

<div style='background:#F0F0F0'>**2** Now repeat the above but apply a Hanning window to the frequency window *before* you apply it to the FFT of $x_n$.  Comment on how well the filter is doing when you do this.  (Its helpful to keep the previous result up)
</div>

```python

w = np.hanning(1000*fc*2)
Xh = np.fft.fft(xn)
Xh[(ff>fc) & (ff<(1./dt-fc))] = 1.e-15
Xh[:fc*1000] = Xh[:fc*1000]*w[fc*1000:]
Xh[-fc*1000:] = Xh[-fc*1000:]*w[:fc*1000]
yh = np.real(np.fft.ifft(Xh))
pyh, fyh = mlab.psd(yh,NFFT=nfft,Fs=1./dt)
px, fx = mlab.psd(xn,NFFT=nfft,Fs=1./dt)
pn, fn = mlab.psd(noise,NFFT=nfft,Fs=1./dt)
Gxyh, f = mlab.csd(xn,yh,NFFT=nfft,Fs=1./dt)

fig,ax = plt.subplots(3,1)
fig.set_size_inches(12,12)
ax[0].plot(t[400:600],xn[400:600], label='signal')
ax[0].plot(t[400:600],y[400:600], label='smoothed-boxcar windowed')
ax[0].plot(t[400:600],yh[400:600], label='smoothed-hanning windowed, fc = %1.2f' %fc)
ax[0].set_xlabel('t [s]')
ax[0].set_ylabel('x [V]')
ax[0].legend()

ax[1].loglog(fn,np.sqrt(py/(px+pn)),'g', label='boxcar')
ax[1].loglog(fyh,np.sqrt(pyh/(px+pn)),'r', label='hanning')
ax[1].set_ylabel('|H(f)|')
ax[1].set_xlabel('f [Hz]')
ax[1].legend()

ax[2].semilogx(f,np.angle(Gxy),'g', label='boxcar')
ax[2].semilogx(f,np.angle(Gxyh),'r', label='hanning')
ax[2].set_ylabel('phase')
ax[2].set_xlabel('f [Hz]')
ax[2].legend()
```

The Hanning window begins to fall off much sooner than the step function.

<div style='background:#F0F0F0'>**3** So the Hanning window falls off a *lot* earlier and more slowly than a step function, so adjust the cutoff frequency to get a reasonable filter that preserves the amplitude of the sine wave at $f=0.03\ \mathrm{Hz}$.  Comment on the pros and cons of this approach.  
</div>

```python
fc = 0.12
w = np.hanning(1000*fc*2)
Xh = np.fft.fft(xn)
Xh[(ff>fc) & (ff<(1./dt-fc))] = 1.e-15
Xh[:fc*1000] = Xh[:fc*1000]*w[fc*1000:]
Xh[-fc*1000:] = Xh[-fc*1000:]*w[:fc*1000]
yh = np.real(np.fft.ifft(Xh))
pyh, fyh = mlab.psd(yh,NFFT=nfft,Fs=1./dt)
px, fx = mlab.psd(xn,NFFT=nfft,Fs=1./dt)
pn, fn = mlab.psd(noise,NFFT=nfft,Fs=1./dt)
Gxyh, f = mlab.csd(xn,yh,NFFT=nfft,Fs=1./dt)

fig,ax = plt.subplots(3,1)
fig.set_size_inches(12,12)
ax[0].plot(t[400:600],xn[400:600], label='signal')
ax[0].plot(t[400:600],y[400:600], label='smoothed-boxcar windowed')
ax[0].plot(t[400:600],yh[400:600], label='smoothed-hanning windowed, fc = %1.2f' %fc)
ax[0].set_xlabel('t [s]')
ax[0].set_ylabel('x [V]')
ax[0].legend()

ax[1].loglog(fn,np.sqrt(py/(px+pn)),'g', label='boxcar')
ax[1].loglog(fyh,np.sqrt(pyh/(px+pn)),'r', label='hanning')
ax[1].set_ylabel('|H(f)|')
ax[1].set_xlabel('f [Hz]')

ax[2].semilogx(f,np.angle(Gxy),'g', label='boxcar')
ax[2].semilogx(f,np.angle(Gxyh),'r', label='hanning')
ax[2].set_ylabel('phase')
```

Asjusting the cutoff frequency for the hanning window improved the quality of the filter, but now we are filtering out less of the high frequency noise. So if the point was to filter out this noise, we loose some of that ability.

## Q2: Implementing a bandpass filter

<div style='background:#F0F0F0'>**1** Create a red-noise process with 20000 data points, $dt=1\ s$.  Using a **truncated Dirichlet Kernel**, create a **bandpass** filter centered at $f_c=0.1\ Hz$ that has a passband that is approximately 0.1 Hz wide. Choose a filter length that is long enough to accomodate 0.1 Hz. 
</div>

   1. Plot the unshifted and shifted filters versus $k$, the time index.
   2. Plot the spectra, the response function $H^2(f)$ and the phase response
   3. Comment on your bandpass and if it indeed passed the right frequencies.

HINT: when you make the Dirichlet Kernel filter make sure that the $k=0$ case is treated "correctly" because $sin(0)=0$ and you will get a discontinuity.  Calculate what $h_0$ should be from [L'Hôpital's_rule](http://en.wikipedia.org/wiki/L'Hôpital's_rule).

```python
N = 20000
t=np.arange(0,N*dt,dt)
x = np.cumsum(np.random.randn(N))
x=x-np.mean(x)
fc = 0.1
mc = 0.1*N
dt = 1.
filtlen = 100
offset = 0.000001
M = fc*2*N+1
k = np.array(range(-N/2+1,N/2))
h = (1./N)*np.sin(np.pi*(k+offset)*M/N)/np.sin(np.pi*(k+offset)/N)
h = h[abs(k)<filtlen+1]
k = k[abs(k)<filtlen+1]
hs = 2*h*np.cos(2*np.pi*(k)*mc/N)

y = np.convolve(x,h,mode='same')
nfft=N/10
pxx,fx=mlab.psd(x,Fs=1.,NFFT=nfft,noverlap=nfft/2)
pyy,fy=mlab.psd(y,Fs=1.,NFFT=nfft,noverlap=nfft/2)
Gxy, fxy = mlab.csd(x,y,NFFT=nfft,Fs=1./dt)

fig,ax = plt.subplots(4,1)
fig.set_size_inches(10,16)
ax[0].plot(k,h,label='unshifted')
ax[0].plot(k,hs,'r',label='shifted')
ax[0].set_xlabel('k')
ax[0].set_ylabel('$h_k$')
ax[0].legend()

ax[1].plot(t[1000:1200],x[1000:1200],label='Raw')
ax[1].plot(t[1000:1200],y[1000:1200],'r',label='Filtered')
ax[1].set_xlabel('t [s]')
ax[1].set_ylabel('x [V]')
ax[1].legend()

ax[2].loglog(fy,pyy/pxx,'r')
ax[2].set_ylabel('$|H(f)|^2 = G_{yy}/G_{xx}$')
ax[2].set_xlabel('f [Hz]')

ax[3].semilogx(fxy,np.angle(Gxy),'r')
ax[3].set_ylabel('phase')
ax[3].set_xlabel('f [Hz]')
```

The bandpass did a good job of filtering frequencies higher than 0.1Hz, although there is some ringing as the filter falls off.
