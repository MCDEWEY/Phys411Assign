### **Q1** Sampling theorem and aliasing

<div style='background:#F0F0F0'>**1** Create a band-limited white noise timeseries with $N=1000$ entries, $dt=0.1 s$, but with all frequencies greater than 0.5 Hz filtered.  Plot the power spectrum to make sure you got it correctly...
</div>

```python
import numpy as np
import matplotlib.mlab as mlab
import scipy.stats as stats
import matplotlib.pyplot as plt
import math 
import cmath
import scipy.signal as signal
%matplotlib inline
```

```python
N = 1000
dt = 0.1
t = np.arange(0,N*dt,dt)
x0 = np.random.randn(N)
B = 0.5
fp = B*0.7
fs = B*0.9
fNyq = 1/2./dt
n,fn=signal.ellipord(fp/fNyq, fs/fNyq,.1, 70.)
b,a=signal.ellip(n,.1,70.,fn,btype='lowpass')
x = signal.filtfilt(b,a,x0)
nfft = np.size(x0)
px0,fx0=mlab.psd(x0,NFFT=nfft,Fs=1./dt,noverlap=nfft/2.,window=mlab.window_hanning)
pxx,fxx=mlab.psd(x,NFFT=nfft,Fs=1./dt,noverlap=nfft/2.,window=mlab.window_hanning)
fig,ax = plt.subplots(1,1)
fig.set_size_inches(14,6)
ax.loglog(fx0,px0,color='0.5')
ax.loglog(fxx,pxx,color='k')
ax.set_xlabel('f [Hz]')
ax.set_ylabel('$G_{xx} [V^2 Hz^{-1}]$')
plt.axvline(x=B,linestyle='--',color='k')
ax.text(B,0.001,' B=%1.1f Hz' %B)
```

<div style='background:#F0F0F0'>**2** Now decimate by a factor of 4 to get a time series sampled at 2.5 Hz.  Show that you can recover the 10 Hz time series using the Sampling Theorem.</div>

```python
xn = x[::4]
tn = t[::4]
fig,ax = plt.subplots(1,1)
fig.set_size_inches(14,6)
ax.plot(t,x,'k')
ax.plot(tn,xn,'*g')
ax.set_xlabel('t [s]')
```

You can see the sampled data set (green stars) does a good job of capturing the variability in the original data. This is because the sampling theorem says we can accurately represent a time series by sampling at a frequency larger than 2B. Here 2B=1Hz, so sampling at 2.5Hz is fine.

<div style='background:#F0F0F0'>**3** Now decimate by a factor of 40 to get a time series sampled at 0.25 Hz.  Show that you cannot recover the 10 Hz time series using the Sampling Theorem.</div>

```python
xn = x[::40]
tn = t[::40]
fig,ax = plt.subplots(1,1)
fig.set_size_inches(14,6)
ax.plot(t,x,'k')
ax.plot(tn,xn,'*g')
ax.set_xlabel('t [s]')
```

Here the sampled time series (green stars) does not accurately capture the variability in the original data. This is because we are samping at 0.25Hz which is less than the minimum frequency 2B=1Hz allowed by the sampling theorem.
