## Q1: Distribution of power spectral estimates

<div style='background:#F0F0F0'>**1** You **may** use `matplotlib.mlab.psd` for the following question, but you had best use it correctly!

Using normally distributed random noise time series of length $N=2048$, show using a Monte Carlo analysis and by comparison to the theoretical probability distribution functions that:  
</div>

   1. The raw spectral estimate is indeed distributed as $\chi^2_2$.
   2. Show that block averaging with no overlap is distributed as $\chi^2_{2N_{blocks}}$.
   3. Show that block averaging with 50% overlap Hanning windows is distributed as $\chi^2_{2N_{blocks}}$.

<div style='background:#F0F0F0'>You can choose your block length, but making it an integer divisor of 2048 will make your life a lot easier.  Show that as $N_{blocks}$ is increased the variance drops (i.e. the distribution of the spectral estimates gets tighter).</div>


<div style='background:#F0F0F0'>HINT: for this time series the individual frequency estimates are indipendent samples of the distribution, so you can use them in compiling your distributions.</div>

<div style='background:#F0F0F0'>HINT: To get the pdf of the $\chi^2_\nu$ distribution correct, you need to "scale" by $\nu^{-1}$ where $\nu$ are the degrees of freedom.</div>

<div style='background:#F0F0F0'>HINT: for good presentation, make your histograms have the same bin sizes, and compare the distributions for all three cases on the same plot.</div>

```python
import numpy.random as random
import numpy as np
import matplotlib.mlab as mlab
import scipy.stats as stats
import matplotlib.pyplot as plt
import math 
import cmath
%matplotlib inline
```

```python
fig,ax = plt.subplots(1,1)
fig.set_size_inches(15,4)

N = 2048.
dt = 1.
nfft = 256

G = np.zeros((N/2,100))
G2 = np.zeros((nfft/2,100))
G3 = np.zeros((nfft/2,100))


for i in range(100):
    x = np.random.randn(N)
    g,f = mlab.psd(x,NFFT=2048,Fs=1./dt)
    g2,f = mlab.psd(x,NFFT=nfft,Fs=1./dt)
    g3,f = mlab.psd(x,NFFT=nfft,Fs=1./dt,window=mlab.window_hanning,noverlap=nfft/2.)
    G[:,i] = g[1:,0]/2
    G2[:,i] = g2[1:]/2
    G3[:,i] = g3[1:]/2
    
G = G.flatten()
G2 = G2.flatten()
G3 = G3.flatten()

ax.hist(G,bins=50,normed=True,histtype='step',label='Monte Carlo of raw spectral estimate')
ax.hist(G2,bins=50,normed=True,histtype='step',label='Monte Carlo with%2.0f bins' %(N/nfft))
ax.hist(G3,bins=50,normed=True,histtype='step',label='Monte Carlo with %2.0f bins' %((2*N/nfft)-1))

y = np.linspace(0,12,1000)
chi1 = stats.chi2.pdf(y,2.,scale=1/2.)
df2 = 2*N/nfft
chi2 = stats.chi2.pdf(y,df2,scale=1./df2)
df3 = 2*(df2-1)
chi3 = stats.chi2.pdf(y,df3,scale=1./df3)
ax.plot(y,chi1,'c',label='$\chi^2_2$')
ax.plot(y,chi2,'k',label='$\chi^2_{%2.0f}$' %df2)
ax.plot(y,chi3,'m',label='$\chi^2_{%2.0f}$' %df3)
ax.legend()
ax.set_ylabel('Normalized Probability Distribution')
ax.set_xlabel('Power Spectrum Gxx $[K^2 Hz^{-1}]$')
ax.set_title('Distribution of Power Spectral Estimates')
```

The raw spectral estimate agrees well with the $\chi^2_2$ distribution, as does the windowed/no-overlap estimate with the $\chi^2_{16}$ distribution, and the Hanning windowed with the $\chi^2_{30}$ distribution. 

As $N_{blocks}$ increases (higher degrees of freedom in $\chi^2$ you can see that the distribution narrows and the variance decreases. This is due to the block averaging. 

<div style='background:#F0F0F0'>**2**  For the Hanning window estimate, what fits better $2N_{blocks}$ degrees of freedom or $18N_{blocks}/11$ degrees of freedom (the correct theoretical value)?  Its OK to evaluate by eye.</div>

```python
fig,bx = plt.subplots(1,1)
fig.set_size_inches(15,4)
bx.hist(G3,bins=50,normed=True,histtype='step',label='Monte Carlo with %2.0f bins' %((2*N/nfft)-1))
bx.plot(y,chi3,'m',label='$\chi^2_{%2.0f}$' %df3)
df4 = (18./11)*(df2-1)
chi4 = stats.chi2.pdf(y,df4,scale=1./df4)
bx.plot(y,chi4,label='$\chi^2_{%2.0f}$' %df4)
bx.legend()
bx.set_ylabel('Normalized Probability Distribution')
bx.set_xlabel('Power Spectrum Gxx $[K^2 Hz^{-1}]$')
bx.set_title('Hanning window estimate with different $\chi^2$ degrees of freedom')
```

The $\chi^2_2$ distribution seems to fit better, although this is not the corrent theoretical value.

## Q2: Spectra of a "red-noise" signal

<div style='background:#F0F0F0'> **1** Load in the Deep Cove hourly data, and compute the power spectra using some reasonable value for $N_{FFT}$, and comment on the effect of applying the Hanning window to not applying it to the spectral leakage.</div>

```python
hourdata=np.genfromtxt('http://web.uvic.ca/~jklymak/Phy411/Data/AllHourly.txt')[[6,28],2:]
dc = hourdata[0,:]
dc=dc[np.isfinite(dc)]+273.15
```

```python
fig,cx = plt.subplots(1,1)
fig.set_size_inches(20,6)
nfft = np.size(dc)/20
dt = 3600.
gnw,fnw = mlab.psd(dc,NFFT=nfft,Fs=1./dt,window=mlab.window_none,noverlap=nfft/2.)
g,f = mlab.psd(dc,NFFT=nfft,Fs=1./dt,window=mlab.window_hanning,noverlap=nfft/2.)
cx.loglog(f[2:],g[2:],label='Power Spectra with Hanning window')
cx.loglog(fnw[1:],gnw[1:],label='Power Spectra without Hanning window')
cx.set_xlabel('f [Hz]')
cx.set_ylabel('Gxx $[K^2 Hz^{-1}]$')
cx.set_title('Power Spectral Density for Deep Cove Temperature Data')
cx.legend()
```

We can see the reduced spectral leakage using the Hanning windowing in the relative smoothness of the blue curve, and in the slightly narrower base of the peaks in the blue curve (most evident at the peak around $10^{-5} Hz$

<div style='background:#F0F0F0'> **2** Now do the same thing to the intergal of the Deep Cove data with time: $$y(t)=\int_o^t x(t')\ \mathrm{d}t'$$  This is obviously a silly thing to do, but compare the spectra and comment on the difference between the Hanning window and the non-Hanning windowed data.  </div>

```python
y  = np.cumsum(dc)
fig,dx = plt.subplots(1,1)
fig.set_size_inches(20,6)
nfft = np.size(dc)/20
dt = 3600.
gnw,fnw = mlab.psd(y,NFFT=nfft,Fs=1./dt,window=mlab.window_none,noverlap=nfft/2.)
g,f = mlab.psd(y,NFFT=nfft,Fs=1./dt,window=mlab.window_hanning,noverlap=nfft/2.)
dx.loglog(f[2:],g[2:],label='Power Spectra with Hanning window')
dx.loglog(fnw[2:],gnw[2:],label='Power Spectra without Hanning window')
dx.set_xlabel('f [Hz]')
dx.set_ylabel('Gxx $[K^2 Hz^{-1}]$')
dx.set_title('Power Spectral Density for Integral of Deep Cove Temperature Data')
dx.legend()
```

Without windowing, none of the features of the original data are preserved in the spectral estimate of the integrated data set. The Hanning windowing has captured more of the underlying variance in the process and has allowed us to resolve some important features.
